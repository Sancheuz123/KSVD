```python
df = pd.read_csv('shopping_trends.csv', encoding= 'latin1')
df
```

```python
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title('Розподіл віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Кількість')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='item_purchased', data=df, order=df['item_purchased'].value_counts().index, palette='viridis')
plt.title('Найпопулярніші товари')
plt.xlabel('Кількість покупок')
plt.ylabel('Назва товару')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='purchase_amount_(usd)', data=df, alpha=0.6)
plt.title('Залежність суми покупок від віку')
plt.xlabel('Вік')
plt.ylabel('Сума покупки')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['review_rating'], fill=True, color='orange')
plt.title('Розподіл оцінок товарів')
plt.xlabel('Рейтинг')
plt.ylabel('Щільність')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_rating', y='purchase_amount_(usd)', data=df, hue='gender', alpha=0.5)
plt.title('Розподіл покупок за статтею')
plt.xlabel('Рейтинг відгуку')
plt.ylabel('Сума покупки')
plt.legend(title='Стать')
plt.show()
```
