```python
import altair as alt
from vega_datasets import data
import plotly.express as px

wheat = data.wheat()
bars = alt.Chart(wheat).mark_bar().encode(
    x='wheat:Q',
    y='year:O'
)

text = alt.Chart(wheat).mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    x='wheat:Q',
    y='year:O',
    text='wheat:Q'
)


chart1 = (bars + text).properties(
    title='Врожайність пшениці за роками',
    width=1000,
    height=700

)

chart1.show()
```

```python
iowa_energy = data.iowa_electricity()

chart2 = alt.Chart(iowa_energy).mark_bar(opacity=0.7).encode(
    x='year:T',
    y='net_generation:Q',
    color='source:N'
).properties(
    width=600,
    title="Виробництво електроенергії в штаті Айова"
)

chart2.display()
```

```python
df = px.data.iris()

fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    size="petal_length",
    color="species",
    hover_data=['petal_width'],
    title="Аналіз параметрів ірисів"
)

fig.show()
```
